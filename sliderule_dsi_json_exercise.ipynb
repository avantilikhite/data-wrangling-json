## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## JSON exercise

Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
normalized_id = json_normalize([json_wb_df],'_id')
normalized_id
```

```python
id_with_countrycode = normalized_id.join(json_wb_df['countrycode'])
id_with_countrycode
```

```python
grouped  = id_with_countrycode.groupby('countrycode')
group_count = grouped.count()
group_count
```

**1. Find the 10 countries with most projects:**

```python
group_count.sort_values('$oid', ascending=False).head(10)
```

** 2. Find the top 10 major project themes (using column 'mjtheme_namecode')**

```python
project_list = json.load((open('data/world_bank_projects.json')))
normalized_mjtheme_namecode = json_normalize(project_list,'mjtheme_namecode',['id','countryname'])
normalized_mjtheme_namecode
```

```python
grouped_theme_ids = normalized_mjtheme_namecode.groupby('code')
grouped_theme_ids.count()
```

```python
grouped_theme_ids['code'].count().sort_values(ascending=False).head(10)
```

**3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.**

```python
normalized_mjtheme_namecode.head()
```

```python
normalized_mjtheme_namecode['name'] = normalized_mjtheme_namecode['name'].replace("", np.nan)
normalized_mjtheme_namecode.head()
```

```python
sorted_deparments = normalized_mjtheme_namecode.sort_values(['code', 'name'])
sorted_deparments.tail()
```

```python
sorted_deparments.fillna(method='ffill', inplace=True)
sorted_deparments.tail()
```

```python
grouped_theme_ids = sorted_deparments.groupby('name')
grouped_theme_ids['name'].count().sort_values(ascending=False).head(10)
```
